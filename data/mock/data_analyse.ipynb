```python
import torch
from loader.txt_processor import load_config, load_data
import pandas as pd

data_cfg = load_config('./data_cfg.txt')
device = torch.device(data_cfg['device'])
```

```python
U, C, T, Y, cost, _, cat1_9 = load_data(data_cfg, isTrainSet=False)
context = [U, cat1_9, T, cost, C, Y]
df = pd.DataFrame(context, index=['User', 'feature', 'TimeStamp','Cost', 'Journey', 'Conversion']).T
df.head()
```

```python
jc = df.iloc[:, -2:]
jc
```

```python
df_1 = df[(df.Conversion==1)]
count_7 = 0
count_12 = 0
count_same = 0
for i, row in df_1.iterrows():
    if len(row['Journey']) >= 7:
        count_7 += 1
    if len(row['Journey']) >= 12:
        count_12 += 1
    if len(set(row['Journey'])) == 1:
        count_same += 1
print('Conversion Journey analyse')
print(f'No.>7:{count_7}\tNo.>12:{count_12}\tNo.Same Num:{count_same}\tNo.Conversion:{len(df_1)}')
```

```python
df_0 = df[(df.Conversion==0)]
count_7 = 0
count_12 = 0
count_same = 0
for i, row in df_0.iterrows():
    if len(row['Journey']) >= 7:
        count_7 += 1
    if len(row['Journey']) >= 12:
        count_12 += 1
    if len(set(row['Journey'])) == 1:
        count_same += 1
print('NonConversion Journey analyse')
print(f'No.>7:{count_7}\tNo.>12:{count_12}\tNo.Same Num:{count_same}\tNo.NonConversion:{len(df_0)}')
```
